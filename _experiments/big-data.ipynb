```python
import byron
```

```python
single_data = byron.f.integer_parameter(0, 2 ** 8)
data = byron.f.sequence([single_data, ','])
blob = byron.f.bunch([data], size=32)
byron.f.as_forest(blob)
```

```python
byron.f.estimate_size(byron.f.bunch([data], size=32 * 1024))
```

```python
data = byron.f.array_parameter(range(2 ** 8), 32 * 1024, sep=', ')
macro = byron.f.macro("{mem}", mem=data)
byron.f.as_forest(macro)
```

```python
byron.f.estimate_size(macro)
```

```python
byron.f.as_text(macro)
```
